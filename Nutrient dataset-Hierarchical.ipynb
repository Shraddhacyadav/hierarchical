```python
import pandas as pd
import numpy as np
import warnings; warnings.simplefilter('ignore')
nutrient = pd.read_csv("G:\\Data\\Datasets\\nutrient.csv",index_col=0)
nutrient.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
nutrientscaled=scaler.fit_transform(nutrient)
# Perform the necessary imports
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

```python
mergings = linkage(nutrientscaled,method='complete')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 12
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size
```

```python
# Plot the dendrogram, using varieties as labels
dendrogram(mergings,
           labels=np.array(nutrient.index),
           leaf_rotation=60,
           leaf_font_size=10,
)

plt.show()
```

```python
from scipy.spatial.distance import pdist
Y = pdist(nutrientscaled, 'mahalanobis', VI=None)

# Calculate the linkage: mergings
mergings = linkage(nutrientscaled,method='complete',metric='mahalanobis')
```

```python
# Plot the dendrogram, using varieties as labels
dendrogram(mergings,
           labels=np.array(nutrient.index),
           leaf_rotation=60,
           leaf_font_size=10,
)
plt.show()
```
